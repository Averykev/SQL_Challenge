```python
#import dependencies

from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from config import password
```

```python
#create an engine for my employee database

engine = create_engine(f"postgresql://postgres:{password}@localhost:5432/Employee_Database_SQL")
conn = engine.connect()
```

```python
#query the information from the salaries table

salaries_data = pd.read_sql("SELECT * FROM salaries", conn)
salaries_data.head()
```

```python
#drop the employee number from the table since I only need the salary info for the histogram

employee_salaries_df = salaries_data.drop(columns = 'emp_no')
employee_salaries_df.head()
```

```python
#plot the salary information with a histogram that has 12 bins

salary_histogram = employee_salaries_df.plot.hist(bins = 12)
```

## NEXT PROBLEM

```python
#query the data from the employees table

title_data = pd.read_sql("SELECT * FROM employees", conn)
title_data.head()
```

```python
#drop the columns that we don't need for our bar chart

title_data = title_data.drop(columns = ['birth_date','sex','hire_date'])
title_data.head()
```

```python
#rename the column so I am able to merge on title_id

title_data = title_data.rename(columns={"emp_title": "title_id"})
title_data.head()
```

```python
#query the data from the title table

titles = pd.read_sql("SELECT * FROM titles", conn)
titles
```

```python
#merge the title table with the employee table

title_merge = pd.merge(title_data, titles, on="title_id", how = "inner")
title_merge
```

```python
#then merge the above table with the salary table so I have all of the information that I need

final_merge = pd.merge(title_merge, salaries_data, on= "emp_no", how = "inner")
final_merge
```

```python
#group the table by title

average_salary = final_merge.groupby("title").mean()
average_salary
```

```python
#drop the employee number from the averages since I don't need it to do my bar chart

average_salary = average_salary.drop(columns = "emp_no")
average_salary
```

```python
#reset the index so I can use the title for the x-axis

average_salary = average_salary.reset_index()
average_salary
```

```python
#plot the average salary information using a bar chart, different colors, and tick locations

x_axis = average_salary['title']
ticks = np.arange(len(x_axis))
y_axis = average_salary['salary']

plt.bar(x_axis, y_axis, align = 'center', color = ['purple','b','g','y','r','m','orange'])

plt.xticks(ticks, x_axis, rotation ='vertical')

plt.ylabel('Average Salary')
plt.xlabel('Employee Title')
plt.title('Average Salary by Employee Title')

plt.show()
```
